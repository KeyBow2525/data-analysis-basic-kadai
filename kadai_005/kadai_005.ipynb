```python
import numpy as np
import matplotlib.pyplot as plt

# 2行3列の計6領域のサブプロットを作成。
fig, axes = plt.subplots(2, 3, tight_layout = True)

# 配列を折れ線で1行1列目に表示。
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

# データを作成。
# 0から10までの幅で配列を等間隔に100個作成。
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

# プロット。
axes[0, 1].plot(x1, y1, linewidth=2.0)

# 軸の設定。
axes[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10), ylim=(-1, 5), yticks=np.arange(-1, 6))

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x2 = np.random .normal(15, 5, 2000)

# ヒストグラムをプロット
axes[0, 2].hist(x2)

# データを生成する
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

# プロットする
axes[1, 0].scatter(x3, y3)

plt.show()
```
